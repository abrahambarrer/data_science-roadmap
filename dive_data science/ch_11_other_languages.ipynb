# Other Languages

```
import pandas as pd
players = pd.read_csv('https://bradfordtuckfield.com/players.csv', encoding = 'ISO-8859-1')
games = pd.read_csv('https://bradfordtuckfield.com/games.csv', encoding = 'ISO-8859-1')
shots = pd.read_csv('https://bradfordtuckfield.com/shots.csv', encoding = 'ISO-8859-1')
```

```
print(players.head())
print(games.head())
print(shots.head)
```

## SQLite3

```
import sqlite3 
conn = sqlite3.connect("soccer.db") 
curr = conn.cursor()
```

```
 players.to_sql('playertable', conn, if_exists='replace', index = False)
```

```
curr.execute('''
SELECT *
FROM playertable
LIMIT 5
''')
```

```
for row in curr.fetchall(): 
    print(row)
```

```
curr.execute(''' 
SELECT * FROM playertable WHERE playerID=554 
          ''') 
for row in curr.fetchall(): 
    print (row)
```

### Joining tables

```
games.to_sql('gamestable', conn, if_exists='replace', index = False) 

curr.execute(''' 
SELECT * FROM gamestable limit 5 
          ''') 

for row in curr.fetchall(): 
    print (row)
```

```
shots.to_sql('shotstable', conn, if_exists='replace', index = False) 
curr.execute(''' 
SELECT * FROM shotstable limit 5 
          ''') 
for row in curr.fetchall(): 
    print (row)
```

```
curr.execute(''' 
SELECT 
    * 
FROM
    shotstable 
    JOIN playertable ON shotstable.shooterID=playertable.playerID 
LIMIT 5 
          ''') 
for row in curr.fetchall(): 
    print(row)
```

## Getting Familiar with R

```
 from rpy2 import robjects
```

```
robjects.r(''' 
my_variable<-512 
print(my_variable+12) 
''')
```

```
robjects.r(''' 
players<-read.csv('https://bradfordtuckfield.com/players.csv') 
print(head(players)) 
''')
```

```
robjects.r(''' 
shots<-read.csv('https://bradfordtuckfield.com/shots.csv') 
print(head(shots)) 
''')
```

```
robjects.r(''' 
print(head(shots$minute)) 
print(head(shots$positionX)) 
''')
```

```
robjects.r(''' 
shot_location_model <- lm(positionX~minute,data=shots) 
print(summary(shot_location_model)) 
''')
```

```
robjects.r(''' 
png(filename='the_plot_chapter11.png') 
plot(shots$minute,shots$positionX) 
abline(shot_location_model) 
dev.off() 
''')
```
