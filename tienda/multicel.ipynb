```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Ánalisis exploratorio

```python
datos_2019 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2019.csv')
datos_2020 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2020.csv')
datos_2021 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2021.csv')
datos_2022 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2022.csv')
datos_2023 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2023.csv')
datos_2024 = pd.read_csv('d:\\cursos\\data_science\\data\\datos_multi_2024.csv')
```

```python
ventas_2019 = datos_2019.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
ventas_2020 = datos_2020.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
ventas_2021 = datos_2021.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
ventas_2022 = datos_2022.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
ventas_2023 = datos_2023.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
ventas_2024 = datos_2024.groupby(['Dia','Fecha','DiaSemana']).sum().drop(['Producto','Cantidad'],axis=1).reset_index()
```
