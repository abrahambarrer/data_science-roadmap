```python
import pandas as pd
import numpy as np
```

# Dataset

Cargando el dataset:

```python
data = pd.read_csv('d:\\cursos\\data_science\\data\\ventas.csv')
# data = pd.read_csv('ventas.csv')
data.head()
```

Eliminando las filas vacías:

```python
data.dropna(how='all',inplace=True)
```

Proceso para rellenar valores NaN según la fecha relativa:

```python
values = data.Fecha.unique()
```

```python
intervals = np.zeros(len(values), dtype='int64')
values = np.delete(values, 1)
```

```python
for i in range(len(values)):
    intervals[i] = data[data.Fecha == values[i]].index.min()
```

```python
intervals[157] = data.index.max()
```

```python
i = 0
for b in range(1,len(intervals[1:158])+1):
    a = b - 1
    data.loc[intervals[a]:intervals[b],'Fecha'] = data.loc[intervals[a]:intervals[b],'Fecha'].fillna(values[i])
    i += 1
```

Eliminar filas con importe 'V'

```python
data = data.replace(['V'],np.nan)
```

Cambiar formato de fecha a legible por Pandas y cambiando formato de Importe a numérico:

```python
data.Importe = pd.to_numeric(data.Importe)
data.Fecha = pd.to_datetime(data.Fecha, format='%d/%m/%Y')
```

Añadiendo columna para día:

```python
data['Dia'] = np.nan
```

```python
i = 0
for j in values:
    data.loc[data.Fecha == j, 'Dia'] = i
    i += 1
```

Añadiendo columna con el nombre del día de la semana:

```python
from datetime import datetime
data['DiaSemana'] = data['Fecha'].apply(lambda fecha: fecha.strftime('%A'))
```

Eliminando columnas innecesarias:

```python
data.drop(['Documento', 'Caja', 'Usuario', 'Total'], axis=1, inplace=True)
```

Reorganizar columnas:

```python
data = data.reindex(columns=['Dia', 'Fecha', 'DiaSemana', 'Cantidad', 'Producto', 'Importe'])
data = data.rename(columns={'Importe': 'Total'})
```
