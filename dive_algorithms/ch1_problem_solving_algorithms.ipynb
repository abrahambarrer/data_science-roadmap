# Problem-Solving

## Analytic approach

$$\begin{align}
x&=v_{0_{x}}t\\
y&=v_{0_{y}}t+\frac{1}{2}at^2
\end{align}$$

Sustituyendo la primera ecuaci√≥n en la segunda en $t$:
$$y=\frac{v_{0_{y}}}{v_{0_{x}}}x+\frac{ax^2}{2v_{0_{x}}^2}$$

```python
def ball_trajectory(x):
    position = 10 * x - 5 * (x**2)
    return(position)
```

```python
import matplotlib.pyplot as plt
xs = [x/100 for x in list(range(201))]
ys = [ball_trajectory(x) for x in xs]
plt.plot(xs, ys)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Trayectoria')
plt.show()
```

## Algorithmic approach

```python
xs2 = [0.1,2]
ys2 = [ball_trajectory(0.1),0]
xs3 = [0.2,2]
ys3 = [ball_trajectory(0.2),0]
xs4 = [0.3,2]
ys4 = [ball_trajectory(0.3),0]
plt.title('Trayectoria con lineas de vision')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(xs,ys,xs2,ys2,xs3,ys3,xs4,ys4)
plt.show()
```
