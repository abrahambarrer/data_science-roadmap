# Computer simulations

El experimento es lanzar 20 monedas y ver cuántas veces sale H.

```python
import numpy as np
```

```python
faces = ['H', 'T']
results = list(np.random.choice(faces, 20))
```

```python
results.count('H') # Resultado de un experimento
```

Ahora, puedes repetir el experimento muchas veces! ¿Cuál es el resultado promedio de 1000 experimentos? O sea, de hacer 1000 experimentos de 20 lanzamientos cada uno.

```python
num_sims = 100
num_flips = 20
results = np.random.binomial(n=num_flips, p=0.5, size=num_sims)
results
```

# Scatter plots

```python
import matplotlib.pyplot as plt
# Only for JupyterLab
# %matplotlib notebook 
```

```python
plt.scatter(range(num_sims), results)
plt.title('Results for 100 experiments flipping a coin')
plt.xlabel('Experiments')
plt.ylabel('Number of heads')
plt.show()
```

# Histograms

```python
min_value = np.unique(results).min()
max_value = np.unique(results).max()
edges = np.arange(min_value-0.5, max_value+1, 1)
plt.hist(results, bins=edges, edgecolor='black', density=True)
# plt.yticks(np.arange(0,22,2))
plt.grid(axis='y')
plt.show()
```

```python
values, counts = np.unique(results, return_counts=True)
values, counts
```

```python
counts / counts.sum()
```

Nótese que aquí ya podemos ver que la proporción de resultados en que H sale 6 veces es muy rara, el 2%!

# Statistical test

```python
def coinsim(num_sims=100, num_flips=20, threshold=6):
    results = np.random.binomial(n=num_flips, p=0.5, size=num_sims)
    print(f'The result of {num_sims} experiments of flipping {num_flips} coins when threshold {threshold}',
        f'is {results[results <= threshold].size / results.size}')

coinsim(1_000_000, 20)
```
